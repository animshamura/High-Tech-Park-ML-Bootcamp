```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
np.random.seed(42)
X_linear = np.linspace(0, 2, 100)  # Keeping the same variable name
y_linear = 0.2 * X_linear**2 - 4 * X_linear + 5 + np.random.randn(100) * 0.3  # New quadratic equation
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_linear, y_linear, alpha=0.7)
plt.title("Data Points")
plt.xlabel("X")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
m = np.random.randn()
c = np.random.randn()

learning_rate = 0.1
num_iterations = 1000

loss_history = []

for i in range(num_iterations):
    y_pred = m * X_linear + c                            # prediction = mx + c ; where x is input data, m is slope and c is y-intercept

    # Calculate the mean squared error loss
    error = y_pred - y_linear                            # Prediction - actual
    mse = np.mean(error ** 2)                            # 1/N SUM[(prediction-actual)^2]       => final MSE loss
    loss_history.append(mse)

    ######## Compute gradients (Calculating gradient of the loss function with respect to m and c)

    # ---- Breakdown (for a single sample) ----------

    # d/dm (prediction-actual)^2 = d/dm [ ( m * X_linear + c ) - y_linear ]^2   => 2 * [ ( m * X_linear + c ) - y_linear ] * X_linear  => 2 * error * X_linear
    # d/dc (prediction-actual)^2 = d/dc [ ( m * X_linear + c ) - y_linear ]^2   => 2 * [ ( m * X_linear + c ) - y_linear ] * 1 => 2 * error
    # Where,  error = y_pred - y_linear  => [ ( m * X_linear + c ) - y_linear ]

    # -------- for all N samples the gradients will be averaged ----------

    dm = 2 * np.mean(error * X_linear)          # Derivative of MSE with respect to m
    dc = 2 * np.mean(error)                     # Derivative of MSE with respect to c

    # Update parameters using gradient descent
    m -= learning_rate * dm                     # M_new = M_old - learning_rate * dm
    c -= learning_rate * dc                     # C_new = C_old - learning_rate * dc

    if i % 100 == 0:
        print(f"Iteration {i}: m = {m:.4f}, c = {c:.4f}, loss = {mse:.4f}")

print("Optimized parameters:")
print("m =", m)
print("c =", c)
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(X_linear, y_linear, label='Data')
plt.plot(X_linear, m*X_linear + c, 'r-', label=f'Best fit: y = {m:.1f}x + {c:.1f}')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Data and Best Fit Line')
plt.legend()
```
